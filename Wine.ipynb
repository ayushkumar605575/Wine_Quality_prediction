# WINE QUALITY PREDICTION

```python
# Importing required Modules
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
```

# Load the DATASET

```python
df = pd.read_csv('winequality-red.csv')
df.head()
```

# DATA PREPROCESSING

```python
for i in range(1599):
    item = df['quality'][i]
    if item <=4:
        df['quality'][i] = 1
    elif item <=6:
        df['quality'][i] = 2
    else:
        df['quality'][i] = 3
```

```python
df['quality'].unique()  # Classification for quality checking
```

```python
df.info()
```

```python
df.describe()
```

```python
# Separate features and target  
data = df.values
X = data[:,0:11]
Y = data[:,11]
```

# MODELLING AND ACCURACY SCORE

```python
# Split the data to train and test dataset.
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,shuffle=True)
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
# Predict from the test dataset
predictions = rfc.predict(X_test)

# Calculate the accuracy
print(f"Accuracy = {accuracy_score(y_test, predictions) * 100}%")
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
# Predict from the test dataset
predictions = dtc.predict(X_test)

# Calculate the accuracy
print(f"Accuracy = {accuracy_score(y_test, predictions) * 100}%")
```

```python
svc = SVC(gamma='auto',kernel='linear')
svc.fit(X_train,y_train)
# Predict from the test dataset
predictions = svc.predict(X_test)

# Calculate the accuracy
print(f"Accuracy = {accuracy_score(y_test, predictions) * 100}%")
```

# SAVING BEST MODEL

```python
from pickle import dump, load
with open("wine_model_RTC.h5",'wb') as model:
    dump(rfc,model)
```

# MODEL PREDICTION

```python
#Prediction of the quality from the input vector
X_new = np.array([[5.9,0.645,0.12,2.0,0.075,32.0,44.0,0.99547,3.57,0.71,10.2]])

quality = {
    1.:'Bad Quality',
    2.:"Fair Quality",
    3.:"Best Quality"
}

with open("wine_model_RTC.h5",'rb') as model:
    model = load(model)

prediction = quality[model.predict(X_new)[0]]

print("Prediction of Wine Quality: {}".format(prediction))
```
